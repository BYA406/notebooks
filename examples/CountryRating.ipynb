```python
import tradingeconomics as te
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

```python
te.login('jupyter:jupyter') # Signup for a free API key at https://developer.tradingeconomics.com/
```

```python
data = te.getHistoricalRatings(country= 'netherlands', output_type='df')['Rating']
data
```

```python
data_counted = Counter(data)
data_counted
```

```python
dates = list(data_counted.keys())
values = list(data_counted.values())

fig1, ax1 = plt.subplots(figsize=(10, 10))

ax1.pie(values, autopct='%1.1f%%',
        startangle=160,
       pctdistance=0.9)

ax1.axis('equal') 

ax1.legend(dates,
          title="Ratings",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1),
          fontsize=10,
          ncol=2,
          title_fontsize=13,
          borderpad=1)


ax1.set_title("Netherlands - Rating Overall", size=20)

plt.show()
```
