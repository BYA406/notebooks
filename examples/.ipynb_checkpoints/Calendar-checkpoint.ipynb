Importing dependencies

```python
import tradingeconomics as te
import configparser as cp
```

Importing data

```python
cfg = cp.ConfigParser()
cfg.read('tradingeconomics.cfg')
key = cfg['API']['key']
te.login(key)

df = te.getCalendarData(output_type='df')
```

Data manipulation

```python
df['Hour'] = df['Date'].apply(lambda x: x[11:])
df['Date'] = df['Date'].apply(lambda x: x[:10])

df = df[['Date', 'Hour', 'Country', 'Category', 'Event', 'Source', 'Actual', 'Previous', 'Forecast', ]]

df.sort_values(['Date', 'Hour'], inplace=True)

df.head(30)
```
