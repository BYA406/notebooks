### Manipulating data

```python
fig, ax = plt.subplots(figsize=(15,10))

x_microsoft = raw_data_microsoft[0:1]['Symbol']
y_microsoft = raw_data_microsoft[0:1]['last']
ax.bar(x_microsoft, y_microsoft, label='Microsoft')

x_apple = raw_data_apple[0:1]['Symbol']
y_apple = raw_data_apple[0:1]['last']
ax.bar(x_apple, y_apple, label='Apple')

x_amazon = raw_data_amazon[0:1]['Symbol']
y_amazon = raw_data_amazon[0:1]['last']
ax.bar(x_amazon, y_amazon, label='Amazon')

x_google = raw_data_google[0:1]['Symbol']
y_google = raw_data_google[0:1]['last']
ax.bar(x_google, y_google, label='Google')




ax.legend()

```
