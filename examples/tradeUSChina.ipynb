```python
import tradingeconomics as te
import pandas as pd
import matplotlib.pyplot as plt


te.login()
```

```python
data = te.getCmtHistorical(symbol = 'USACHN27091', output_type='df')

dates = pd.DatetimeIndex(data['date'])
labels = dates.year
values = data['value']

%matplotlib inline
plt.style.use('ggplot')

x = labels
crude = values
x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos,crude, width=0.8, bottom=None, align='center', data=None,  color='black')
plt.xlabel("year")
plt.ylabel("USD Million")
plt.title("US Imports from China of Crude Oil")
plt.ylim(10)
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
plt.xticks(x_pos, x)

plt.show()
```
