Importing dependencies

```python
import tradingeconomics as te
import time
import matplotlib.pyplot as plt
```

Importing data

```python
te.login('10D377002D63488:6E8511E3FE804EC')

population_in_cities = te.getWBIndicator(series_code = 'EN.URB.MCTY.TL.ZS', output_type='df')
time.sleep(1)

covid_cases = te.getIndicatorData(country='all', indicators = 'coronavirus cases', output_type='df')
```

Treating data

```python
population_in_cities = population_in_cities[['last', 'country']].rename(columns={'last': 'Population in big cities (% of total)'})
population_in_cities = population_in_cities.rename(columns={'country': 'Country'}).set_index('Country')

covid_cases = covid_cases[['Country', 'LatestValue']].rename(columns={'LatestValue': 'Cases (millions)'}).set_index('Country')
covid_cases['Cases (millions)'] = covid_cases['Cases (millions)'].apply(lambda x: x/1000000)

df = covid_cases.join(population_in_cities, lsuffix='_covid_cases', rsuffix='_population_in_cities')
# df = df[['Population in big cities (% of total)', 'Cases (millions)']]

df.sort_values(['Cases (millions)', 'Population in big cities (% of total)'], ascending=False, inplace=True)
df.drop('World', inplace=True)
df = df.dropna().head(20)
df
```

Plotting

```python
color_dict = {'Population in big cities (% of total)': 'orange', 'Cases (millions)': 'grey'}

df.plot(
    figsize=(20, 9),
    kind='bar',
    zorder=3,
    color=[color_dict.get(x, '#333333') for x in df.columns],
    fontsize=12
)

ax = plt.gca()

plt.setp(
    ax.get_xticklabels(),
    rotation=45,
    ha='right',
    rotation_mode='anchor',
    fontsize=12
)

# ax.labelsize=40
plt.yticks(list(plt.yticks()[0]) + [2.5, 5, 7.5, 15])
ax.yaxis.grid(zorder=0)
ax.set_title('Coronavirus Cases x Population in Cities', color='black', size=20)
plt.xlabel('Country', fontsize=16)
plt.tight_layout()
```
