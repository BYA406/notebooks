# Zimbabwe Trade Partners

```python
import tradingeconomics as te
from wordcloud import WordCloud
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
```

```python
te.login('jupyter:jupyter')
```

```python
trading_partners = te.getCmtCountry(country = 'Zimbabwe', output_type='df')
trading_partners
```

```python
# imports
import_partners = trading_partners.loc[trading_partners['type'] == 'Import']
import_partners
```

```python
import_countries = import_partners['country2'].tolist()
import_countries
```

```python
text = " ".join(i for i in import_countries)
wordcloud = WordCloud(background_color="white").generate(text)
plt.figure( figsize=(15,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# exports
export_partners = trading_partners.loc[trading_partners['type'] == 'Export']
export_partners
```

```python
export_countries = export_partners['country2'].tolist()
export_countries
```

```python
text = " ".join(i for i in export_countries)
wordcloud = WordCloud(background_color="white").generate(text)
plt.figure( figsize=(15,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
